```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
House_Price = pd.read_csv("House Price dataset.csv", sep=",")
House_Price.head(10)
```

```python
House_Price.info()
```

```python
from sklearn.preprocessing import LabelEncoder
Label = LabelEncoder()
feature = ["Location",	"Condition",	"Garage"]
for col in feature:
    House_Price[col] = Label.fit_transform(House_Price[col])
    
House_Price
```

```python
House_Price1 = House_Price.drop("Id", axis = 1)
House_Price1
```

```python
y = House_Price1["Price"]
x = House_Price1.drop("Price", axis =1)
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size =20, random_state = 20)
```

```python
reg = linear_model.LinearRegression()
reg.fit(x_train,y_train)

pred_y = reg.predict(x_test)
```

```python
#coefficient, MSE,Variance score

print("Coefficient:\n", reg.coef_)
print("Mean Squared Error:%.2f" % mean_squared_error(y_test,pred_y))
print("Variance score:%.2f" % r2_score(y_test,pred_y))

```
